```python
import pandas as pd
import glob

arquivos = glob.glob("dados/*.csv")

dfs = [pd.read_csv(arq) for arq in arquivos]

df = pd.concat(dfs, ignore_index=True)

print(df.shape)
print(df.head())
```

```python
X = df.select_dtypes(include=['int64', 'float64'])

print(X.columns)
```
